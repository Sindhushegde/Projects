### Data Augmentation ###

Data augmentation is important to improve the performance of the neural network model. Data augmentation techniques help to improve the robustness of the model by forming new and different varients of dataset. This technique is useful to  avoid the overfitting of the data as well.


Transforms are used for image transformation. We can augment the data using transform.compose in order to avoid the overfitting of the data.

### Datasets and DataLoaders ###

PyTorch provides two data primitives: ***torch.utils.data.DataLoader*** and ***torch.utils.data.Dataset*** that allow you to use pre-loaded datasets as well as your own data. Dataset stores the samples and their corresponding labels, and DataLoader wraps an iterable around the Dataset to enable easy access to the samples.

```python
class CustomImageDataset(Dataset):
    def __init__(self, X, y, transform=None, target_transform=None):
        self.X = X
        self.y = y
        self.transform = transform
        self.target_transform = target_transform
    
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, idx):
        image = torch.from_numpy(self.X[idx])
        label = self.y[idx]
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        return (image, label)
```

### Define the model ###

In PyTorch, the idiom for defining a model is to create a class that extends the Module class.

The constructor of the class defines the model layers and forward() function is an override that specifies how to forward propagate input via these layers.

In init function define each layer we used in the model.

Defining the model contains below steps:


* Apply convolution layers - These are importatant to create the feature mapping. This model contains 4 convolution layers[3].
* Introduce the Poolings - 
MaxPooling is used to reduce dimensionality. This is a pooling operation that calculates the maximum, or largest, value in each patch of each feature map.
  **kernel_size**  the size of the window to take a max over - it is 2 in this model [1]

  64x64 image tensors is pooled 4 times with a kernel size of 2. i.e 64/2/2/2/2 = 4, so our feature tensors are now 4x4. And we have generated 48 of them (out channels for the last layer)

*   Flatten the data in order to have a np.array to feed the NN - we need to flatten the data in order to feed them to a fully-connected layer.  
 
 *torch.nn.Linear(in_features, out_features)*

 in_features  size of each input sample

 out_features  size of each output sampl

 We flatten the data by applying linear transformation to   the incoming data[2].


*   forword() function - In the forward function, pass the data through the layers we defined in the init function.

 Used ReLu actiuvation function.

 To avoid overfitting, we have a layer that drops 30% of the features at random.

References: 

[1] https://pytorch.org/docs/stable/generated/torch.nn.MaxPool2d.html
[2] https://pytorch.org/docs/stable/generated/torch.nn.Linear.html
[3]https://www.kaggle.com/androbomb/using-cnn-to-classify-images-w-pytorch

```python

class CNN(nn.Module):
    
    # Defining the Constructor
    def __init__(self, num_classes=3):
        super(CNN, self).__init__()
        
        
        # We will apply 12 filters in the first convolutional layer
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=12, kernel_size=3, stride=1, padding=1)
        
        # A second convolutional layer takes 12 input channels, and generates 24 outputs
        self.conv2 = nn.Conv2d(in_channels=12, out_channels=24, kernel_size=3, stride=1, padding=1)
        
        # A third convolutional layer takes 24 input channels, and generates 36 outputs
        self.conv3 = nn.Conv2d(in_channels=24, out_channels=36, kernel_size=3, stride=1, padding=1)

        # A third convolutional layer takes 36 input channels, and generates 48 outputs
        self.conv4 = nn.Conv2d(in_channels=36, out_channels=48, kernel_size=3, stride=1, padding=1)

        self.pool = nn.MaxPool2d(kernel_size=2)
        
        # below code drops 30% of the features to prevent overfitting
        self.drop = nn.Dropout2d(p=0.3)
        
        # Flatten
        self.fc = nn.Linear(in_features=4 * 4 * 48, out_features=num_classes)

    def forward(self, x):
       
        x = F.relu(self.pool(self.conv1(x))) 
        x = F.relu(self.pool(self.conv2(x)))  
        x = F.relu(self.pool(self.conv3(x)))   
        x = F.relu(self.pool(self.conv4(x)))
        
        # Selecting some feature to drop to prevent overfitting
        x = F.dropout(self.drop(x), training=self.training)
        
        # Flatten
        x = x.view(-1, 4 * 4 * 48)
        # Feed to fully-connected layer to predict class
        x = self.fc(x)
        # Return class probabilities via a log_softmax function 
        return torch.log_softmax(x, dim=1)
    
device = "cpu"
if (torch.cuda.is_available()):
    # if GPU available, use cuda (on a cpu, training will take a considerable length of time!)
    device = "cuda"

# Create an instance of the model class and allocate it to the device
model = CNN(num_classes=len(classes)).to(device)

print(model)
```

### Training the model ###

The training process consists of an iterative sequence of forward passes in which the layers in the network process the training data in batches, and the optimizer changes the weights. We'll also use a different collection of test images to put the model to the test at the end of each epoch, so we can watch how well it does as the training goes on.

Model training is done in below steps[1]:


*   Setting the model to training mode
*   Process the images in batches - In batches, we'll iterate through the files. Inside each batch we must 
  - import labels and features
  - Push the data forward through the layers of the mode
     *model.forword()*
  - compute the loss
  - Backpropagate
     
      *loss.backward()*


*   During each Epoch compute the Average Loss of the Model


*zero_grad()* sets gradients of all model parameters to zero. This helps in decreasing the errors(losses)[2].
